```python
import requests
import matplotlib.pyplot as plt
import numpy as np 
import time

def iss_loc():
    output = requests.get('http://api.open-notify.org/iss-now.json')
    location = output.json()
    longitude = float(location['iss_position']['longitude'])
    latitude = float(location['iss_position']['longitude'])
    return longitude, latitude

long = []
lat = []

for i in range(100):
    longitude, latitude = iss_loc()
    long.append(longitude)
    lat.append(latitude)
    time.sleep(10)
    
x = np.array(long)
y = np.array(lat)

plt.scatter(x, y)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("International Space Station Path")
plt.show()
```
